```python
#this code allows you to calculate the angle of movement between any 2 points
import math
prev_x = 0
prev_y = 0
for j, row in df.iterrows():
        x0 = prev_x
        y0 = prev_y
        x1 = row['easting']
        y1 = row['northing']
#         print(math.degrees(math.atan((y1-y0)/(x1-x0))))
        if y1- y0 == 0 or x1-x0 == 0:
            continue
        df.loc[j, 'angle'] = math.degrees(math.atan((y1-y0)/(x1-x0)))
        prev_x = row['easting']
        prev_y = row['northing']
df
```

```python
seq = Sequential()

seq.add(ConvLSTM2D(filters=40, kernel_size=(3, 3), input_shape=(None, 40, 40, 3), padding='same', return_sequences=True))

seq.add(BatchNormalization())

seq.add(ConvLSTM2D(filters=40, kernel_size=(3, 3), padding='same', return_sequences=True))

seq.add(BatchNormalization())

seq.add(ConvLSTM2D(filters=40, kernel_size=(3, 3), padding='same', return_sequences=True))

seq.add(BatchNormalization())

seq.add(ConvLSTM2D(filters=40, kernel_size=(3, 3), padding='same', return_sequences=True))

seq.add(BatchNormalization())

seq.add(Conv3D(filters=1, kernel_size=(3, 3, 3),activation='sigmoid', padding='same', data_format='channels_last'))

seq.compile(loss='categorical_crossentropy', optimizer='adadelta')
```
